```python
#get coordinates
import re
all_coordinates = re.findall(r"N\d{0,2}\W\W\d{0,2}' \d{0,2}\", W\d{0,3}\W\W\d{0,3}' \d{0,3}\"", formatted_document)

print(len(all_coordinates))
print(all_coordinates)
```

```python
#convert latitude and longitude degrees minutes seconds to decimal values. This will only work in Northern Hemisphere!!! 
import re

def dms2dd(degrees, minutes, seconds):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    return dd;

def north_dms2dd(degrees, minutes, seconds):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    dd*= -1;
     
    return dd;

def dd2dms(deg):
    d = int(deg)
    md = abs(deg - d) * 60
    m = int(md)
    sd = (md - m) * 60
    return [d, m, sd]

def parse_dms(dms):
    parts = re.split('[^\d{0,2}]+', dms)
    lat = dms2dd(parts[1], parts[2], parts[3])
    lng = north_dms2dd(parts[5], parts[6], parts[7])

    return [lat, lng]

converted_coordinates = []

for i in all_coordinates: 
    converted_coordinate = [parse_dms(i)]
    converted_coordinates += converted_coordinate
#     return converted_coordinate

print(converted_coordinates[42])
print(converted_coordinates)
```
